```
import glob
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import PIL
import PIL.Image
import tensorflow as tf
import tensorflow_datasets as tfds
import keras.preprocessing.image
from keras.preprocessing.image import ImageDataGenerator, load_img,array_to_img, img_to_array
from PIL import Image

artistNames = ["Alfred_Sisley","Amedeo_Modigliani","Camille_Pissarro","Diego_Velazquez",
               "Edgar_Degas","Edouard_Manet","Edvard_Munch","Frida_Kahlo","Georges_Seurat",
               "Gustav_Klimt","Henri_de_Toulouse-Lautrec","Henri_Matisse","Henri_Rousseau"]
artistNames = []
artistNames = artistNames + ["Kazimir_Malevich","Marc_Chagall","Mikhail_Vrubel","Pablo_Picasso"]
artistNames = artistNames + ["Paul_Cezanne","Pierre-Auguste_Renoir","Piet_Mondrian","Pieter_Bruegel"]
artistNames = artistNames + ["Rene_Magritte","Salvador_Dali","Vasiliy_Kandinskiy","Vincent_van_Gogh","William_Turner"]
    
categories = ["impressionist"]

DATADIR = "/content/drive/MyDrive/CS 230 Group Project/impressionist"

augmentor = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

training_data = []


    


```

```
def augment(dim):
  for category in artistNames:
    print(category)
    path = os.path.join(DATADIR,category)
    r = 0
    for img_name in os.listdir(path):
      r+=1
      #print(r)
      print(path + '/'+ img_name)
      image = load_img(path + '/'+ img_name)
      image = image.resize((dim, dim), Image.ANTIALIAS)
      image_arr = img_to_array(image)
      image_arr = image_arr.reshape((1,)+image_arr.shape)
      i = 0
      for batch in augmentor.flow(image_arr,batch_size=1,
                                save_to_dir='/content/drive/MyDrive/CS 230 Group Project/color',
                                save_prefix='pre',
                                save_format='jpeg'):
        i = i + 1
        if i>=10:
          break
```
